## Univariate Exploration

> In this section, investigate distributions of individual variables. If
you see unusual points or outliers, take a deeper look to clean things up
and prepare yourself to look at relationships between variables.

```python

bins = np.arange(0, loan['MonthlyLoanPayment'].max()+100, 100)
plt.hist(data = loan, x = 'MonthlyLoanPayment',bins=bins)
plt.title('Monthly Payment')
plt.xlabel('Payment ($)');
```

```python
#we have a long tail so we will use log scale


bins_log = 10 ** np.arange(2, np.log10(loan['MonthlyLoanPayment'].max())+0.03, 0.03)
plt.hist(data = loan, x = 'MonthlyLoanPayment',bins=bins_log)
plt.title('Monthly Payment (log-scale)')
plt.xlabel('Payment ($)')
plt.xscale('log')
plt.xticks([100,150 , 300, 600, 1000,2000], [100,150 , 300, 600, '1k','2k'])
plt.show();

```


we have a long tail distrbution which means most people tend to pay as low as they can for monthly payment. when ploted a log scale looks roughly bimodal with one near to 150 and the other near 300. We can see after the first peak there is a hug jump and then hug drop.

## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

```python
plt.scatter(data = loan, y = 'BorrowerRate', x = 'MonthlyLoanPayment', alpha = 0.2)
plt.title('Monthly Payment Vs Interset Rate')
plt.xlabel('Monthly Payment ($)')
plt.ylabel('Interest Rate');
```

```python
plt.hist2d(data = loan, y = 'BorrowerRate', x = 'MonthlyLoanPayment',cmap = 'viridis_r', cmin = 0.5)
plt.title('Monthly Payment Vs Interset Rate')
plt.xlabel('Monthly Payment ($)')
plt.ylabel('Interest Rate');
```

Lookng at the scutter plot we can see that the lower the interst rate the higher the monthly payment goes but because of the overlapping of observation points we can't conclude anything when the monthly payment goes down.

In the heat map it is clear that when the interest rate increases the monthly payment decreases.

```python
base_color = sb.color_palette()[0]
sb.violinplot(data = loan, x = 'MonthlyLoanPayment', y = 'IncomeRange', color = base_color,
              inner = None);
plt.title('Monthly Payment Vs Income Range')
plt.xlabel('Monthly Payment ($)')
plt.ylabel('Income Range');
```

```python
sb.violinplot(data = loan, y = 'BorrowerRate', x = 'ProsperScore', color = base_color,
              inner = None);
plt.title('Prosper Score Vs Monthly Payment')
plt.ylabel('Interest rate')
plt.xlabel('Prosper Score');
```

We can see from the Heatmap when the interest rate deacreases the Monthly payment Increases.

 We can see from the (Monthly Payment Vs Income Range) plot that costumer with higher income have tend to pay higher monthly payment.
Also, from the other plot we can see clearly that cotumer with higher Prosper Score get lower interest rate
